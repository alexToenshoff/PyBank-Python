# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_return_path = Path("Resources/whale_returns.csv")
wr_df = pd.read_csv(whale_return_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
wr_df.sort_index(inplace=True)
wr_df.head()
```

```python
# Count nulls
wr_df.isnull().sum()
```

```python
# Drop nulls
wr_df = wr_df.dropna().copy()
wr_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
ar_path = Path("Resources/algo_returns.csv")
ar_df = pd.read_csv(ar_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
ar_df.sort_index(inplace=True)
ar_df.head(7)
```

```python
# Count nulls
wr_df.isnull().sum()
ar_df.isnull().sum()
```

```python
# Drop nulls
wr_df = wr_df.dropna().copy()
ar_df = ar_df.dropna().copy()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
# Reading algorithmic returns
sp500_path = Path("Resources/sp500_history.csv")
sp500_df = pd.read_csv(sp500_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_df.sort_index(inplace=True)
sp500_df.head()
```

```python
# Check Data Types
print(sp500_df.dtypes)
print(ar_df.dtypes)
print(wr_df.dtypes)
```

```python
# Fix Data Types
sp500_df["Close"]= sp500_df["Close"].str.replace("$", "")
sp500_df["Close"]= sp500_df["Close"].str.replace(",", "")
sp500_df["Close"]= sp500_df["Close"].astype("float")
print(sp500_df.dtypes)
```

```python
# Calculate Daily Returns
sp500_dr = sp500_df.pct_change()
sp500_dr
```

```python
# Drop nulls
sp500_dr.dropna()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_dr = sp500_dr.rename(columns={"Close" : "SP500 Daily Return"})
sp500_dr.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_df = pd.concat([wr_df, ar_df, sp500_dr], axis="columns", join="inner")
combined_df.head()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratio = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
sharpe_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio.plot(kind='bar')
```

```python
# Reading data from 1st stock
g_path = Path("Resources/goog_historical.csv")
googl_df = pd.read_csv(g_path, na_values='NaN', keep_default_na=False, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
googl_df.head()
```

```python
# Reading data from 2nd stock
cost_path = Path("Resources/cost_historical.csv")
cost_df = pd.read_csv(cost_path, na_values='NaN', keep_default_na=False, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
cost_df.head()
```

```python
# Reading data from 3rd stock
aaplpath = Path("Resources/aapl_historical.csv")
aapl_df = pd.read_csv(aaplpath, na_values='NaN', keep_default_na=False, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
aapl_df.head()

sp500_path = Path("Resources/sp500_history.csv")
sp500_df = pd.read_csv(sp500_path, na_values='NaN', keep_default_na=False, index_col="Date", infer_datetime_format=True, parse_dates=True)
```

```python
# Combine all stocks in a single DataFrame
combined_df = pd.concat([aapl_df, googl_df, cost_df, sp500_df], axis="columns", join="inner")
combined_df.drop(columns=["Symbol"], inplace=True)
combined_df.columns = ["AAPL NOCP", "GOOG NOCP", "COST NOCP", "SP500 NOCP"]
combined_df.head()
```

```python
# Reset Date index
#already done^^
print(combined_df.dtypes)
```

```python
# Reorganize portfolio data by having a column per symbol
##already done^^
```

```python
# Calculate daily returns
mppc = combined_df.pct_change()[1:]

# Drop NAs
mppc.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# Display sample data
mppc.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
w = [1/3, 1/3, 1/3]

# Calculate portfolio return
wr = (w * mppc)
wr1 = wr.sum(axis=1)

# Display sample data
wr1.head(10)
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
data_with_wei = pd.concat([mppc, wr1], axis="columns", join="inner")
dww = data_with_wei
columns = ["AAPL % Change", "GOOG % Change", "COST % Change", "Weighted Average % Change"]
dww.columns = columns
dww.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
dww.dropna()
```
